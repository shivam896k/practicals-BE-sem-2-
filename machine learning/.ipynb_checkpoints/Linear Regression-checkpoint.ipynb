```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Training data
hours = [[10], [9], [2], [15], [10], [16], [11], [16]]
risk = [95, 80, 10, 50, 45, 98, 38, 93]
```

```python
# Scatter plot to explore data
plt.scatter(hours, risk)
plt.xlabel('Number of hours spent in driving')
plt.ylabel('Risk score on scale of 0-100')
plt.show()
```

```python
# Fitting regression line through data
model = LinearRegression()
model.fit(hours, risk)
print(model.coef_)
print(model.intercept_)
```

```python
# Plotting regression line through data
y = model.coef_*hours + model.intercept_
plt.plot(hours, y)
plt.scatter(hours, risk)
plt.xlabel('Number of hours spent in driving')
plt.ylabel('Risk score on scale of 0-100')
plt.show()
```
